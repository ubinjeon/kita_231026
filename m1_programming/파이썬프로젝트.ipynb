- 예상 작업 manhour 보다 실제 더 많이 소요된 경우, 평가지표 및 더 빨리 일을 처리할 수 있는 개선되어야 하는 방안 확인
- 물류 사업 단가가 합당한지 아닌지. 즉 사업성이 있는지 판단할 수 있음.
- 본사나 분석팀만 쓰는 것이 아니라 실제 작업하는 인원들도 쓰면서 업무 효율에 적극적으로 개선하는데 참여하도록 도와주는 프로그램.
- 작업들은 연속적이다. 한 가지 작업으로 이루어져있지 않다. 여러 단계(단위)의 작업들을 세분화한 뒤 연속해서 합산. 예)준비작업->테스트->실제작업->정리
- 현장에서 이루어지는 작업들을 눈으로 볼 수 있어서 사무실이나 커뮤니케이션이 필요한 곳에서 보고 이해하는데 용이.
- 물류 증가가 예상되는 월에 대한 팀구성 및 필요 manhour 산출 가능

<프로그램 내에 반영되어져야 하는 내용들>
- 필요 업무 포지션 예) forklift driver, picker, packer, loader, unloader, supervisor etc)
- 작업 스페이스의 효율의 정도
- 평균 (대략) 창고 내 이동거리: -> 포크리프트가 창고 안에서 다니는 평균 속도가 있으면 이동 시간이 나옴.
- 제품 종류: 한가지 혹은 여러가지
	- 같은 종류에 대한 given 작업량 처리 시간에 대한 산출기
	- 여러 많은 종류에 대한 given 작업량 처리
- 팀 수 및 팀구성
- 업무 어려움 정도

<가능한 Extension>
- 창고의 환경이 다 다르다.
- 제품이 다 다르다.
- 경험. 즉 처음 할 때와 점점 익숙해지면 걸리는 시간이 다르다.

- 프로젝트 매니지먼트랑 연계할 수 있음. 프로젝트 타임라인 구성시.

<Input -> Process -> Output>

    - difficulty: 1=비교적 작업하기 쉬움, 2=중간, 3=비교적 어려움 (깨지기 쉽거나 위험하여 다루기 어려움)  
    - shape: 1=regular, 2=long or big, 세변의 합이 150cm 를 넘는가
    - weight: 1: <5kg, 2: <20kg,3 >=20kg

- Process:
    - defined 한 창고 
    - 프로젝트에서는 위의 3가지 작업 특성을 조합해서 18개의 작업 종류를 정의
    - 데이터셋 generation: 18가지의 고유의 작업단위 (업무 어려움 정도랑 효율의 정도는 고유의 속성)에 대하여
        - 각 작업단위에 필요한 팀 구성
        - 각 포지션의 필요 맨hour range 정의
        - random하게 
    
- Output: 각 포지션의 필요 맨hour 산출. 견적서 형태로.

```python
## 작업의 정의 - 최소단위의 작업 즉 같은 작원 팀의 작업 예) 1000개의 시트를 검사하는데 필요한 
# 작업단위 이름 주기 <- class 로 만들까?
# 이 작업을 수행하기 위해 필요한 업무 포지션들 추가
# 이 작업을 수행하기 위해 필요한 최소 인원

## 작업 효율의 정도 및 랜덤하게 데이터셋을 뽑아내어서 
# 작업 스페이스의 효율의 정도 <- 이것은 뺐음
# 평균 (대략) 창고 내 이동거리
# 한가지 아이템에 대한 것인지. 즉 단순 반복인지

from itertools import product
import random
import pandas as pd
#작업을 정의하고 작업을 업무 어려움과 효율의 정도(한 예로 이동거리)에 따라서 분류하기
#work_title: str, team(position & 필요인원): dict,
#difficulty(업무 어려움의 척도 즉 3이 제일 어려움): 1,2,3
# def define_workunit (work_title, team, difficulty):
# #
# '''
# <Position>
# Supervisor -> 최대 인원 1 # 평균 시간 단위는 하루에 0.1 단위로 0.5~2 랜덤
# Forklift driver -> 최대 인원 2 # 평균 시간 단위는 하루에 0.1 단위로 0.5~5 랜덤
# Picker -> 최대 인원 5 # 평균 시간 단위는 하루에 0.1 단위로 2~7 랜덤
# Packer -> 최대 인원 2 # 평균 시간 단위는 하루에 0.1 단위로 2~7 랜덤
# '''
# position 과 인원, efficiency, difficulty의 모든 조합
#     supervisor_outcome = [0,1] # Work_definition의 인덱스 0
#     forkliftdriver_outcome = [0,1] # Work_definition의 인덱스 1
#     picker_outcome = [1,2,3,4] # Work_definition의 인덱스 2
#     packer_outcome = [1,2] # Work_definition의 인덱스 3
#     difficulty_outcome = [1,2,3] # Work_definition의 인덱스 4
#     work_definition = list(product(*[supervisor_outcome,forkliftdriver_outcome,picker_outcome,packer_outcome,difficulty_outcome]))
#     print(f"경우의 수 : {len(work_definition)}개") #위의 수로 180개의 업무단위 조합
#     print(work_definition)
    
# # 업무 어려움 척도
# def difficulty_coefficient ()
# # 각 케이스별로 평균적인 어려움 coefficient
# # difficulty_coefficient 180개에 대한 각 position의 coefficient 더블리스트 있어야 함. <- 이것은 데이터를 만든다음 그것을 돌리면서 산출

# work_title: str, difficulty: int (1,2,3), shape: int (1=regular, 2=long or big, 세변의 합이 150cm 를 넘는가), 
# weight: int(1: <1kg,2: <10kg,3 <30kg, workload: int
def define_workunit (work_title, difficulty, shape, weight, workload):
    
    #작업을 어려움, 아이템모양, 아이템무게를 조합해서 18개의 작업 종류를 정의
    difficulty_outcome = [1,2,3] # Work_definition의 각 튜플 요소의 인덱스 0
    shape_outcome = [1,2] # Work_definition의 각 튜플 요소의 인덱스 1
    weight_outcome = [1,2,3] # Work_definition의 각 튜플 요소의 인덱스 2
    work_definition = list(product(*[difficulty_outcome,shape_outcome,weight_outcome]))
    print(f"경우의 수 : {len(work_definition)}개") #위의 수로 18개의 업무단위 조합
    print(work_definition)

def team_formation(work_definition):
#     supervisor_outcome = [0,1] # Work_definition의 인덱스 0
#     forkliftdriver_outcome = [0,1] # Work_definition의 인덱스 1
#     picker_outcome = [1,2,3,4] # Work_definition의 인덱스 2
#     packer_outcome = [1,2] # Work_definition의 인덱스 3
    supervisor_num = [0]*18 #18가지의 업무단위 조합에 필요한 팀의 supervisor 수. 기본값: 0
    forkliftdriver_num = [0]*18 #기본값: 0
    picker_num = [1]*18 #기본값: 1
    packer_num = [1]*18 #기본값: 1
    
# 각 작업단위 특성을 고려하여 팀의 각 포지션별 필요 수 정하기
    for i in range(len(work_definition)):
        # 팀 내의 supervisor 수: 전체적인 업무강도 정도의 합이 4가 넘을 경우 +1
        if sum(work_definition [i]) > 4:
            supervisor_num[i]= 1
         # 팀 내의 forkliftdriver 수: 아이템 모양이 크거나 길때 즉 2일때 혹은 아이템이 30kg 이상으로 무거울 때 즉 3일때 +1    
        if (work_definition [i][1] > 1 or work_definition [i][2] > 2):
            forkliftdriver_num[i] = 1
        # 팀 내의 picker 수:
        ## weight = 3, shape = 2, difficulty =3 인경우 각각  picker수 +1
        if work_definition[i][2] == 3:
            picker_num[i] += 1
        if work_definition[i][1] == 2:
            picker_num[i] += 1
        if work_definition[i][0] == 3:
            picker_num[i] += 1
        # 팀 내의 packer 수:
        ## shape = 2, weight= 3 인경우 각각 packer수 +1 
        if work_definition[i][1] == 2:
            packer_num[i] += 1
        if work_definition[i][2] == 3:
            packer_num[i] += 1

# data를 work_definition
def data_generation(work_definition, supervisor_num,forkliftdriver_num, picker_num, packer_num):
#    data = {'work_index':[],'work_load':[],'s_mh': [], 'f_mh':[], 'pi_mh':[], 'pa_mh':[]}
    column_name = ['work_index','workload','s_mh', 'f_mh', 'pi_mh', 'pa_mh']
    work_index = []*18000
    workload = []*18000
    s_mh = []*18000
    f_mh = []*18000
    pi_mh = []*18000
    pa_mh = []*18000
    
    # random range로 쓸 작업단위 별 포지션의 한 사람당 예상 작업 range
    s_mh_range = [0]*18 #기본값: 0
    f_mh_range = [0] *18 #기본값: 0
    pi_mh_range = [[0.3,0.5]]*18 #기본값: 0.3~0.5 manhour
    pa_mh_range = [[0.3,0.5]]*18 #기본값: 0.3~0.5 manhour
    for i in range(len(work_definition)):
        # 18가지 작업단위에 대한 supervisor의 한사람당 걸릴 것 같은 작업 시간 range:
        # 필요인원이 0인 경우 0
        #if difficulty = 3 or (shape = 2 and weight = 3) supervisorMH range 0.4,0.7
        #else = 0.2~0.4
        if s_mh_range[i] == 0:
            pass
        
        elif work_definition[i][0] == 3 or (work_definition[i][1] == 2 and work_definition[i][3]):
            s_mh_range[i] = [0.4,0.7]
        else:
            s_mh_range[i] = [0.2,0.4]
        
        # forklift manhour range:
        # 필요인원이 0인 경우 0
        #if difficulty = 3 or shape = 2. forklift_MH_range 0.4,0.8 
        #else = 0.2~0.4
        if f_mh_range[i] == 0:
            pass
        
        elif work_definition[i][0] == 3 or work_definition[i][1] == 2:
            f_mh_range[i] = [0.4,0.8]
        else:
            f_mh_range[i] = [0.2,0.4]
        
        # picker manhour range:
        # if difficulty >1 or shape = 2 or weight >1, 0.5~0.8
        # else 기본값
        if work_definition[i][0] >1 or work_definition[i][1] == 2 or work_definition[i][2] >1:
            pi_mh_range[i] = [0.5,0.8]
            
        # packer manhour range:
        # if difficulty >1 or shape = 2 or weight >1, 0.5~0.8
        # else 기본값
        if work_definition[i][0] >1 or work_definition[i][1] == 2 or work_definition[i][2] >1:
            pa_mh_range[i] = [0.5,0.8]

    # 18개의 작업단위에 대하여 각각 10000개의 workload 1~100의 랜덤정수,
    # 's_mh', 'f_mh', 'pi_mh', 'pa_mh'는 바로 위에 정의한 range에 맞춰서 랜덤float 곱하기
    # workload<5 1.5를 곱하기 *workload*s_num*random *가중치 sum(work_definition[work_index])/10
    for case in range(1000):
        for i in range(len(work_definition)):
            work_index[case] = i
            workload[case] = random.randint(1,50)
            s_mh[case] = (1 if s_mh_range[i] == 0 else random.uniform(s_mh_range[i][0],s_mh_range[i][1]) *
                          (1.5 if work_load[case] > 5 else 1) *
                          (sum(work_definition[i])/10) *
                          supervisor_num[i])
            f_mh[case] = (1 if f_mh_range[i] == 0 else random.uniform(f_mh_range[i][0],f_mh_range[i][1]) *
                          (1.5 if work_load[case] > 5 else 1) *
                          (sum(work_definition[i])/10) *
                          forkliftdriver_num[i])
            pi_mh[case] = (random.uniform(pi_mh_range[i][0],pi_mh_range[i][1]) *
                          (1.5 if work_load[case] > 5 else 1) *
                          (sum(work_definition[i])/10) *
                          picker_num[i])
            pa_mh[case] = (random.uniform(pa_mh_range[i][0],pa_mh_range[i][1]) *
                          (1.5 if work_load[case] > 5 else 1) *
                          (sum(work_definition[i])/10) *
                          packer_num[i])
            
    # pandas로 dataset으로 변환하기
    # 인덱스와 열 이름 지정
#pd.DataFrame(zip(a, b), index = pd.Index(['aaa', 'bbb', 'ccc', 'ddd']), columns = column_name)
    df = pd.DataFrame(zip(work_index, workload, s_mh, f_mh, pi_mh, pa_mh), columns = column_name)
    print(df.head())
    return df

#인풋 데이터프레임을 토대로 각 케이스 별 평균 맨파워 도출
def mean_manhour(self, df):
    s_mh_mean = [0]*18
    f_mh_mean = [0]*18
    pi_mh_mean = [0]*18
    pa_mh_mean = [0]*18
    for i in range (18):
        s_mh_mean[i] = df.loc[df['work_index'] == i,'s_mh'].mean()
        f_mh_mean[i] = df.loc[df['work_index'] == i,'s_mh'].mean()
        pi_mh_mean[i] = df.loc[df['work_index'] == i,'s_mh'].mean()
        pa_mh_mean[i] = df.loc[df['work_index'] == i,'s_mh'].mean()
        
    return s_mh_mean, f_mh_mean, pi_mh_mean, pa_mh_mean


def main_output(self):
    print(
f"""
                                   견 적 서
                           
- 작업 아이템: {self.worktitle}
- 주업무: 피킹/패킹
- 업무량: {self.workload}

- 필요 manhour:
    - Supervisor: {self.s_mh_mean} hours
    - Forklift Driver: {self.f_mh_mean} hours
    - Picker: {self.pi_mh_mean} hours
    - Packer: {self.pa_mh_mean} hours
""")

```

```python
from itertools import product
import random
import pandas as pd
import numpy as np
import datetime
import time

class Quotation_Programmer:
    # difficulty: int (1,2,3), shape: int (1=regular, 2=long or big, 세변의 합이 150cm 를 넘는가), 
    # weight: int(1: <1kg,2: <10kg,3 <30kg, workload: int
    #작업을 어려움, 아이템모양, 아이템무게를 조합해서 18개의 작업 종류를 정의
    difficulty_outcome = [1,2,3] # Work_definition의 각 튜플 요소의 인덱스 0
    shape_outcome = [1,2] # Work_definition의 각 튜플 요소의 인덱스 1
    weight_outcome = [1,2,3] # Work_definition의 각 튜플 요소의 인덱스 2
    work_definition = list(product(*[difficulty_outcome,shape_outcome,weight_outcome]))

    def team_formation(self):
#     supervisor_outcome = [0,1]
#     forkliftdriver_outcome = [0,1]
#     picker_outcome = [1,2,3,4]
#     packer_outcome = [1,2]
        self.supervisor_num = [0]*len(self.work_definition) #18가지의 업무단위 조합에 필요한 팀의 supervisor 수. 기본값: 0
        self.forkliftdriver_num = [0]*len(self.work_definition) #기본값: 0
        self.picker_num = [1]*len(self.work_definition) #기본값: 1
        self.packer_num = [1]*len(self.work_definition) #기본값: 1
    
    # 각 작업단위 특성을 고려하여 팀의 각 포지션별 필요 수 정하기
        for i in range(len(self.work_definition)):
            # 팀 내의 supervisor 수: 전체적인 업무강도 정도의 합이 4가 넘을 경우 +1
            if sum(self.work_definition [i]) > 4:
                self.supervisor_num[i]= 1
             # 팀 내의 forkliftdriver 수: 아이템 모양이 크거나 길때 즉 2일때 혹은 아이템이 30kg 이상으로 무거울 때 즉 3일때 +1    
            if (self.work_definition [i][1] > 1 or self.work_definition [i][2] > 2):
                self.forkliftdriver_num[i] = 1
            # 팀 내의 picker 수:
            ## weight = 3, shape = 2, difficulty =3 인경우 각각  picker수 +1
            if self.work_definition[i][2] == 3:
                self.picker_num[i] += 1
            if self.work_definition[i][1] == 2:
                self.picker_num[i] += 1
            if self.work_definition[i][0] == 3:
                self.picker_num[i] += 1
            # 팀 내의 packer 수:
            ## shape = 2, weight= 3 인경우 각각 packer수 +1 
            if self.work_definition[i][1] == 2:
                self.packer_num[i] += 1
            if self.work_definition[i][2] == 3:
                self.packer_num[i] += 1
        return self.supervisor_num, self.forkliftdriver_num, self.picker_num, self.packer_num
    
    def data_generation(self):
        #team_formation의 함수를 사용하여 team 구성하기
        self.team_formation_result = self.team_formation()
        self.supervisor_num = self.team_formation_result[0]
        self.forkliftdriver_num = self.team_formation_result[1]
        self.picker_num = self.team_formation_result[2]
        self.packer_num = self.team_formation_result[3]

        #Data Output 칼럼명 및 칼렴
        self.column_name = ['work_index','workload','s_mh/worklaod', 'f_mh/workload', 'pi_mh/workload', 'pa_mh/workload']
        self.work_index = [0] * len(self.work_definition) * 1000
        self.workload = [0] * len(self.work_definition) * 1000
        self.s_mh = [0] * len(self.work_definition) * 1000
        self.f_mh = [0] * len(self.work_definition) * 1000
        self.pi_mh = [0] * len(self.work_definition) * 1000
        self.pa_mh = [0] * len(self.work_definition) * 1000
        # random range로 쓸 작업단위 별 포지션의 한 사람당 예상 작업 range
        self.s_mh_range = [0]*len(self.work_definition) #기본값: 0
        self.f_mh_range = [0] *len(self.work_definition) #기본값: 0
        self.pi_mh_range = [[0.3,0.5]]*len(self.work_definition) #기본값: 0.3~0.5 manhour
        self.pa_mh_range = [[0.3,0.5]]*len(self.work_definition) #기본값: 0.3~0.5 manhour
        
        # 18가지 작업단위 에 대한 각 업무 포지션 별 한사람당 걸릴 것 같은 작업 시간 range 설정
        for i in range(len(self.work_definition)):
            
            # supervisor manhour range:
            # 필요인원이 0인 경우 0
            #if difficulty = 3 or (shape = 2 and weight = 3) supervisorMH range 0.4,0.7
            #else = 0.2~0.4
            if self.supervisor_num[i] == 0:
                pass

            elif self.work_definition[i][0] == 3 or (self.work_definition[i][1] == 2 and self.work_definition[i][2]):
                self.s_mh_range[i] = [0.4,0.7]
            else:
                self.s_mh_range[i] = [0.2,0.4]

            # forklift manhour range:
            # 필요인원이 0인 경우 0
            #if difficulty = 3 or shape = 2. forklift_MH_range 0.4,0.8 
            #else = 0.2~0.4
            if self.forkliftdriver_num[i] == 0:
                pass

            elif self.work_definition[i][0] == 3 or self.work_definition[i][1] == 2:
                self.f_mh_range[i] = [0.4,0.8]
            else:
                self.f_mh_range[i] = [0.2,0.4]

            # picker manhour range:
            # if difficulty >1 or shape = 2 or weight >1, 0.5~0.8
            # else 기본값
            if self.work_definition[i][0] >1 or self.work_definition[i][1] == 2 or self.work_definition[i][2] >1:
                self.pi_mh_range[i] = [0.5,0.8]

            # packer manhour range:
            # if difficulty >1 or shape = 2 or weight >1, 0.5~0.8
            # else 기본값
            if self.work_definition[i][0] >1 or self.work_definition[i][1] == 2 or self.work_definition[i][2] >1:
                self.pa_mh_range[i] = [0.5,0.8]

        # 18개의 작업단위에 대하여 각각 10000개의 workload 1~100의 랜덤정수,
        # 's_mh', 'f_mh', 'pi_mh', 'pa_mh'는 바로 위에 정의한 range에 맞춰서 랜덤float 곱하기
        # workload<5 1.05를 곱하기 *s_num*random *가중치 sum(work_definition[work_index])/10
        for case in range(1000):
            for i in range(len(self.work_definition)):
                self.work_index[i+(case*18)] = i
                self.workload[i+(case*18)] = random.randint(1,50)
                self.s_mh[i+(case*18)] = (0 if self.s_mh_range[i] == 0 
                                           else random.uniform(self.s_mh_range[i][0],self.s_mh_range[i][1]) *
                                          (1.05 if self.workload[i+(case*18)] > 5 else 1) *
                                          (sum(self.work_definition[i])/10) *
                                          self.supervisor_num[i])
                self.f_mh[i+(case*18)] = (0 if self.f_mh_range[i] == 0 
                                          else random.uniform(self.f_mh_range[i][0],self.f_mh_range[i][1]) *
                                          (1.05 if self.workload[i+(case*18)] > 5 else 1) *
                                          (sum(self.work_definition[i])/10) *
                                          self.forkliftdriver_num[i])
                self.pi_mh[i+(case*18)] = (random.uniform(self.pi_mh_range[i][0],self.pi_mh_range[i][1]) *
                                          (1.05 if self.workload[i+(case*18)] > 5 else 1) *
                                          (sum(self.work_definition[i])/10) *
                                          self.picker_num[i])
                self.pa_mh[i+(case*18)] = (random.uniform(self.pa_mh_range[i][0],self.pa_mh_range[i][1]) *
                                          (1.05 if self.workload[i+(case*18)] > 5 else 1) *
                                          (sum(self.work_definition[i])/10) *
                                          self.packer_num[i])

        # pandas로 dataset으로 변환한 후 csv로 출력. dataset을 리턴
        self.df = pd.DataFrame(zip(self.work_index, self.workload, self.s_mh, self.f_mh, self.pi_mh, self.pa_mh), columns = self.column_name)
        #print(self.df.head())
        self.df.to_csv('test1.csv')
        return self.df
    
    #인풋 데이터프레임을 토대로 각 케이스 별 평균 맨파워 도출
    def mean_manhour(self):
        self.df = self.data_generation()
        self.s_mh_mean = [0]*18
        self.f_mh_mean = [0]*18
        self.pi_mh_mean = [0]*18
        self.pa_mh_mean = [0]*18
        for i in range (18):
            self.s_mh_mean[i] = self.df.loc[self.df['work_index'] == i,'s_mh/worklaod'].mean()
            self.f_mh_mean[i] = self.df.loc[self.df['work_index'] == i,'f_mh/workload'].mean()
            self.pi_mh_mean[i] = self.df.loc[self.df['work_index'] == i,'pi_mh/workload'].mean()
            self.pa_mh_mean[i] = self.df.loc[self.df['work_index'] == i,'pa_mh/workload'].mean()

        return self.s_mh_mean, self.f_mh_mean, self.pi_mh_mean, self.pa_mh_mean

#Quotation_Programmer 상속
class Quotation(Quotation_Programmer):
    
    def __init__(self):
        self.worktitle = input("견적 받고 싶은 아이템에 대하여 8가지 질문을 드릴려고 합니다.\n최대한 정확한 정보를 입력해주시면 감사하겠습니다.\n\nPick/Packing 작업 견적을 원하시는 아이템 이름을 입력하세요 >> ")
        while 1:
            self.fragile = input("입력하신 아이템이 깨지기 쉬운 유리/그릇 류 입니까? (y/n) >> ")
            try:
                if (self.fragile.lower() == 'y' or self.fragile.lower() == 'n'):
                    break
                else:
                    print("잘못입력하셨습니다. y 혹은 n 으로 입력해주세요.")
            except:
                print("잘못입력하셨습니다. y 혹은 n 으로 입력해주세요.")
        while 1:
            self.dangerous = input("입력하신 아이템이 위험화물인 배터리, 향수, 알코올, 화학약품 류 입니까? (y/n) >> ")
            try:
                if (self.dangerous.lower() == 'y' or self.dangerous.lower() == 'n'):
                    break
                else:
                    print("잘못입력하셨습니다. y 혹은 n 으로 입력해주세요.")
            except:
                print("잘못입력하셨습니다. y 혹은 n 으로 입력해주세요.")
        while 1:
            self.easy = input("입력하신 아이템이 가벼운 플라스틱, 금속, 종이 류 입니까? (y/n) >> ")
            try:
                if (self.easy.lower() == 'y' or self.easy.lower() == 'n'):
                    break
                else:
                    print("잘못입력하셨습니다. y 혹은 n 으로 입력해주세요.")
            except:
                print("잘못입력하셨습니다. y 혹은 n 으로 입력해주세요.")
        while 1:
            self.size = input("입력하신 아이템의 가로+세로+높이 합이 160cm을 넘거나 또는 한변의 길이가 100cm 이상입니까? (y/n) >> ")
            try:
                if (self.size.lower() == 'y' or self.size.lower() == 'n'):
                    break
                else:
                    print("잘못입력하셨습니다. y 혹은 n 으로 입력해주세요.")
            except:
                print("잘못입력하셨습니다. y 혹은 n 으로 입력해주세요.")
        while 1:
            try:
                self.weight_kg = int(input("입력하신 아이템의 개당 무게를 숫자(정수)로 입력하세요. (단위: kg) >> "))
                break
            except:
                print("숫자로 입력해주세요.")
        while 1:
            try:
                self.workload_user = int(input("입력하신 작업 아이템에 대한 작업 갯수를 숫자로 입력하세요. (단위: pcs) >> "))
                break
            except:
                print("숫자로 입력해주세요.")
        self.remark = input("그 외 작업 혹은 아이템에 대한 비고사항을 자유롭게 기재하세요 >> ")
        print("입력해주셔서 감사합니다. 견적서를 출력합니다.")
        
        #작업 아이템이 깨지기 쉽거나 작업하기 힘들면 3 작업하기 보통이면 2 작업하기 간단하면 1
        if (self.fragile == 'y' or self.dangerous == 'y'):
            self.difficulty = 3
        elif self.easy == 'y':
            self.difficulty = 1
        else:
            self.difficulty = 2
         #작업 아이템이 정형화된 박스 모양이면 1 부피가 크거나 길어서 세변의 합이 160cm를 넘으면 2
        if self.size == 'y':
            self.shape = 2
        else:
            self.shape = 1
        #작업 아이템이 20kg이상이면 3, 작업 아이템이 20kg 미만 5kg 이상이면 2, 5kg 미만이면 1
        if self.weight_kg >= 20:
            self.weight = 3
        elif self.weight_kg >= 5:
            self.weight = 2
        else:
            self.weight = 1
        
        # work_definition 의 작업 경우의 수 중에 어떤 것인지 튜플 형식으로 묶고 work_definition 상 어떤 인덱스인지
        self.work_index_user = self.work_definition.index((self.difficulty, self.shape, self.weight))
        self.meandata = self.mean_manhour()
        self.result_s = round(self.meandata[0][self.work_index_user].item()* self.workload_user,1)
        self.result_f = round(self.meandata[1][self.work_index_user].item()* self.workload_user,1)
        self.result_pi = round(self.meandata[2][self.work_index_user].item()* self.workload_user,1)
        self.result_pa = round(self.meandata[3][self.work_index_user].item()* self.workload_user,1)

    def print_quotation(self):
        print(
f"""
----------------------------------------------------------------------------------------
                                        견 적 서


- 견적날짜: {datetime.date.today()}

- 작업 아이템: {self.worktitle}
- 주업무: 피킹/패킹
- 아이템 모양: {"박스모양의 정형화된 아이템" if self.shape == 1 else "크거나 긴 비정형화된 아이템"}
- 아이템 무게/개당: {self.weight}kg
- 아이템 종류: {("위험화물인 배터리, 향수, 알코올, 화학약품 류" if self.dangerous == 'y' 
                else "깨지기 쉬운 유리/그릇 류" if self.fragile == 'y'
                else "가벼운 플라스틱, 금속, 종이 류" if self.easy == 'y'
                else "작업 강도 중간")}
- 아이템 개수: {self.workload_user}개
- 작업 혹은 아이템에 대한 비고사항: 
    - {self.remark}

- 필요 예상 manhour:
    - Supervisor: {self.result_s} hours
    - Forklift Driver: {self.result_f} hours
    - Picker: {self.result_pi} hours
    - Packer: {self.result_pa} hours
    
----------------------------------------------------------------------------------------

""")

# 견적서 loop
def main():
    end_botton = 1
    while end_botton:
        a = Quotation() # Quotation 객체 생성 후 Quotation 정보 입력
        time.sleep(1)
        a.print_quotation() # Quotation 클래스의 프린트 함수
        time.sleep(1)
        while 1:
            another = input("또 다른 견적서를 원하시면 y 를, 프로그램 종료를 원하시면 n 을 입력하세요 >> ")
            try:
                if another.lower() == 'y':
                    break
                elif another.lower() == 'n':
                    print("프로그램을 종료합니다. 이용해주셔서 감사합니다.")
                    end_botton = 0
                    break
                else:
                    print("잘못입력하셨습니다. y 또는 n 으로 입력해주세요.")
            except:
                print("잘못입력하셨습니다. y 또는 n 으로 입력해주세요.")
        

main()
```
