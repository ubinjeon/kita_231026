#### [과제1_1103]

파이썬으로 음식 메뉴 추천 프로그램 작성하기

요구 사항:

- 사용자에게 기본 제공되는 메뉴 목록을 보여줍니다.
- 사용자가 추가하고 싶은 메뉴를 입력할 수 있도록 요청합니다.
- 사용자가 "stop"이라고 입력하면 메뉴 입력을 중단합니다.
- 메뉴 목록을 무작위로 섞습니다.
- 사용자가 원하는 추천 메뉴의 갯수를 입력할 수 있도록 합니다.
- 입력한 숫자만큼 메뉴에서 무작위로 추천해줍니다.
- 사용자가 추천 메뉴에 만족하지 않으면 다시 추천할 수 있는 옵션을 제공합니다.

시나리오:

기본으로 제공하는 메뉴:<br>
{'맥도날드', '쉑쉑버거', '편의점', '분식집', '닭곰탕', '짬뽕'}

주변 음식점 이름을 적으세요. 없으면 'stop'을 입력하세요: 중국집<br>
주변 음식점 이름을 적으세요. 없으면 'stop'을 입력하세요: stop<br>
추천받고 싶은 메뉴의 갯수를 말하세요: 2<br>
추천 메뉴는 ['짬뽕', '편의점'] 입니다.<br>

다른 추천을 원하시나요? (y/n): y<br>
추천 메뉴의 수를 2으로 설정합니다.<br>
추천 메뉴는 ['편의점', '닭곰탕'] 입니다.

```python
import random
menu = {'맥도날드', '쉑쉑버거', '편의점', '분식집', '닭곰탕', '짬뽕'}

print(f"""
--------------------------------------------------------------

*********************기본 제공 메뉴 목록**********************
{menu}

--------------------------------------------------------------
""")
# 메뉴 추가
while 1:
    menu_input = input(f"주변 음식점 이름을 적으세요. 없으면 'stop'을 입력하세요: ")
    if menu_input == 'stop':
        break
    else:
        menu.add(menu_input)
        print(f"메뉴가 성공적으로 추가되었습니다.\n<메뉴목록>\n{menu}")
        
# 메뉴 추천
end_chr = 1 # 1==메뉴 추천 루프를 실행, 0==메뉴 추천 루프를 종료
while end_chr == 1:
    menu_num = 0
    while 1:
        menu_num_str = input("추천받고 싶은 메뉴의 갯수를 말하세요 (숫자만): ")
        if menu_num_str.isdigit() == False:
            print("잘못 입력하셨습니다. 숫자만 입력해주세요")
            continue
        if int(menu_num_str) > len(menu):
            print("잘못 입력하셨습니다. {menu.len()} 이하의 숫자만 입력해주세요")
            continue
        menu_num = int(menu_num_str)
        break
    while 1:
        rec_menu = random.sample(sorted(menu),menu_num)
        print(f'추천 메뉴는 {rec_menu} 입니다.')
        diff_menu_yn = input("\n다른 추천을 원하시나요? (y/n) ")
        if diff_menu_yn == "y":
            for i in rec_menu:
                menu.remove(i)
            if len(menu) ==0:
                print("더이상 추천할 수 있는 메뉴가 없습니다.")
                end_chr = 0
                break
            menu_num = random.randint(1,len(menu))
            print(f"추천 메뉴의 수를 {menu_num}으로 설정합니다.")
        elif diff_menu_yn == "n":
            print("메뉴 추천을 종료합니다.")
            end_chr = 0
            break
        else:
            print("죄송합니다. 잘못입력하셨으므로 메뉴 추천을 종료합니다.")
            end_chr = 0
            break
```

#### [과제2_1103] : 메뉴 추천 프로그램을 사용자 함수 3개를 이용해서 구조화 하세요. (add_menu_items, get_recommendation, main)

```python
import random

# 인풋은 추가할 메뉴(str), 메뉴리스트(set)
def add_menu_items(item,menu):
    menu.add(item)
    return menu

# 인풋은 메뉴리스트(set), 추천수(숫자)
def get_recommendation(menu,recommend_num):
    rec_menu = random.sample(sorted(menu),recommend_num)
    return rec_menu

    
def main():
    menu = {'맥도날드', '쉑쉑버거', '편의점', '분식집', '닭곰탕', '짬뽕'}
    # 메뉴 추가
    while 1:
        menu_input = input(f"주변 음식점 이름을 적으세요. 없으면 'stop'을 입력하세요: ")
        if menu_input == 'stop':
            break
        else:
            menu = add_menu_items(menu_input,menu)
            print(f"메뉴가 성공적으로 추가되었습니다.\n<메뉴목록>\n{menu}")

    # 메뉴 추천
    end_chr = 1 # 1==메뉴 추천 루프를 실행, 0==메뉴 추천 루프를 종료
    while end_chr == 1:
        menu_num = 0
        while 1:
            menu_num_str = input("추천받고 싶은 메뉴의 갯수를 말하세요 (숫자만): ")
            if menu_num_str.isdigit() == False:
                print("잘못 입력하셨습니다. 숫자만 입력해주세요")
                continue
            if int(menu_num_str) > len(menu):
                print("잘못 입력하셨습니다. {menu.len()} 이하의 숫자만 입력해주세요")
                continue
            menu_num = int(menu_num_str)
            break
        while 1:
            rec_menu = get_recommendation(menu,menu_num)
            print(f'추천 메뉴는 {rec_menu} 입니다.')
            diff_menu_yn = input("\n다른 추천을 원하시나요? (y/n) ")
            if diff_menu_yn == "y":
                for i in rec_menu:
                    menu.remove(i)   # 중복 추천을 피할 수 있게 추천하는 메뉴 삭제
                if len(menu) ==0:
                    print("더이상 추천할 수 있는 메뉴가 없습니다.")
                    end_chr = 0
                    break
                menu_num = random.randint(1,len(menu))
                print(f"추천 메뉴의 수를 {menu_num}으로 설정합니다.")
            elif diff_menu_yn == "n":
                print("메뉴 추천을 종료합니다.")
                end_chr = 0
                break
            else:
                print("죄송합니다. 잘못입력하셨으므로 메뉴 추천을 종료합니다.")
                end_chr = 0
                break
                
main()
```

```python
# 사용자 함수 활용

import random

def add_menu_items(menu):
  print(f'기본으로 제공하는 메뉴:\n{menu}\n')
  while True:
    user_input = input("주변 음식점 이름을 적으세요. 없으면 'stop'을 입력: ").strip()
    if user_input.lower() == 'stop':
      break
    else:
      menu.add(user_input)
  return list(menu)
def get_recommendation(menu, num_items):
  try:
    picks = random.sample(menu, num_items)
    print(f'추천 메뉴는 {picks} 입니다.\n')
  except ValueError:
    print(f'숫자가 너무 큽니다. 메뉴는 {len(menu)}개밖에 없습니다.')

def main():
  menu = {"짬뽕", "분식집", "맥도날드", "쉑쉑버거", "편의점", "닭곰탕"}
  menu = add_menu_items(menu)
  while True:
    try:
      num_recommendations = int(input('추천받고 싶은 메뉴의 개수를 입력하세요: '))
      get_recommendation(menu,num_recommendations)
      break
    except ValueError:
      print(f'숫자를 적어주세요')
  choice = input('다른 추천을 원하시나요? (y/n): ').strip().lower()
  if choice == 'y':
    recmd_num = random.randrange(1,len(menu)+1)   
    print(f'제가 추천하는 메뉴수는 다음과 같이 {recmd_num}개 입니다')
    get_recommendation(menu,recmd_num)
  elif choice == 'n':
    print('식사 맛있게 하세요~')
  else:
    print('잘못된 입력입니다. 프로그램 종료합니다.')

if __name__ == "__main__":
  main()
```

```python
def page_num(total_contents,num_on_page):
    return ((total_contents//num_on_page) if total_contents%num_on_page==0 else (total_contents//num_on_page +1))

total_contents = int(input("게시물의 총 건수를 입력하세요. "))
num_on_page = int(input("한 페이지에 보여 줄 게시물 수를 입력하세요. "))

page_num(total_contents,num_on_page)
```

#### [과제9_1103] : 파이썬의 random 모듈을 사용하여 간단한 구매 데이터 세트를 생성하세요. 이 데이터 세트에는 다음과 같은 정보가 포함:

- 구매 ID
- 구매자 ID
- 구매한 제품의 ID
- 구매한 제품의 수량
- 구매 날짜와 시간
- purchase_id: 1000에서 9999 사이의 랜덤 정수
- buyer_id: 100에서 999 사이의 랜덤 정수
- product_id: 10에서 99 사이의 랜덤 정수
- quantity: 1에서 10 사이의 랜덤 정수
- purchase_date: 최근 365일 이내의 랜덤 날짜 및 시간

```python
import random
import datetime
import time


with open('purchase_data.txt','w') as file:
    for i in range(1000):
        
        purchase_id = random.randint(1000,9999) #중복될 수 있다는 것을 감안 못했네... 
        buyer_id = random.randint(100,999)
        product_id = random.randint(10,99)
        quantity = random.randint(1,10)

        delta = datetime.timedelta(days=365)
        start_date = datetime.date.today() - delta
        purchase_date_unformatted = random.uniform(time.mktime(start_date.timetuple()),time.time())
        purchase_date = datetime.datetime.fromtimestamp(purchase_date_unformatted)
        purchase_date_informat = purchase_date.strftime('%Y-%m-%d %X')
        
        file.write(f'purchase_id:{purchase_id}, buyer_id:{buyer_id}, product_id:{product_id}, quantity:{quantity}, purchase_date:{purchase_date_informat}\n')

with open('purchase_Data.txt') as file:
    print_data = file.read()
    print(print_data)
```

```python
import random
import time
from datetime import datetime, timedelta

# 데이터 세트의 갯수 입력
data_num = 100

def random_date():
    # 365일 Unix 시간은 31536000 초
    ran_timestamp = random.uniform(0, 31536000)
    ran_time = time.time() - ran_timestamp
    return (datetime.fromtimestamp(ran_time)).strftime('%Y-%m-%d %H:%M:%S')

purchase_id = list(random.randint(1000,9999) for i in range(data_num))
buyer_id = list(random.randint(100,999) for i in range(data_num))
product_id = list(random.randint(10,99) for i in range(data_num))
quantity = list(random.randint(1,10) for i in range(data_num))
purchase_date = list(random_date() for i in range(data_num))

data_set = list(zip(purchase_id, buyer_id, product_id, quantity, purchase_date))

# 데이터 세트 중에서 10개만 출력
for i, data in enumerate(data_set):
    if i == 10: break
    print(data)
```
